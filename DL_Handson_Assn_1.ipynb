# Linear Regression using TF

```python
import tensorflow as tf
import numpy as np
import datetime

# Generate synthetic data: y = 3x + 2 + noise
np.random.seed(0)
x_vals = np.random.rand(100).astype(np.float32)
y_vals = 3 * x_vals + 2 + np.random.normal(scale=0.1, size=100)
```

```python
# Convert to tensorflow tensors
x = tf.constant(x_vals, dtype=tf.float32)
y = tf.constant(y_vals, dtype=tf.float32)

# Initialize trainable weights / Learnable parameters
W = tf.Variable(0.0)
b = tf.Variable(0.0)
```

```python
# Define hyperparameters
learning_rate = 0.1
epochs = 200

# Set up TensorBoard log directory
log_dir = "logs/linreg/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
writer = tf.summary.create_file_writer(log_dir)
```

```python
def compute_gradients(x):
    with tf.GradientTape() as tape:
        y_pred = W * x + b
        loss = tf.reduce_mean(tf.square(y_pred - y))  # MSE
    # Compute gradients
    gradients = tape.gradient(loss, [W, b])
    return gradients, loss

for epoch in range(epochs):
    grad,loss = compute_gradients(x)
    # Update weights manually
    W.assign_sub(learning_rate * grad[0])
    b.assign_sub(learning_rate * grad[1])

    # Log to TensorBoard
    with writer.as_default():
        tf.summary.scalar("loss", loss, step=epoch)
        tf.summary.scalar("W", W, step=epoch)
        tf.summary.scalar("b", b, step=epoch)

    if epoch % 20 == 0:
        print(f"Epoch {epoch:03d}: Loss = {loss:.4f}, W = {W.numpy():.4f}, b = {b.numpy():.4f}")

print(f"Final W = {W.numpy()}, b = {b.numpy()}")
```

### Exercise 2
Implement a binary classifier using logistic regression trained with gradient descent.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
x_vals = np.random.rand(100).astype(np.float32)
y_vals = (x_vals > 0.5).astype(np.float32)  # label = 1 if x > 0.5

# Visualize
plt.scatter(x_vals, y_vals)
plt.title("Binary classification data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

x = tf.constant(x_vals)
y = tf.constant(y_vals)
```

## Equations Needed
$$\hat{y} = \sigma(Wx + b)$$

Where:
- $\sigma(z) = \frac{1}{1 + e^{-z}}$ (sigmoid activation)
- $W, b$: trainable parameters  
- $\hat{y}$: predicted probability

### Loss Function: Binary Cross Entropy

$$\mathcal{L}(y, \hat{y}) = -\frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$$

### Gradient
$\frac{\partial \mathcal{L}}{\partial W},\quad \frac{\partial \mathcal{L}}{\partial b}$

### Update rules for each epoch:
$$W \leftarrow W - \eta \cdot \frac{\partial \mathcal{L}}{\partial W}$$
$$b \leftarrow b - \eta \cdot \frac{\partial \mathcal{L}}{\partial b}$$

Where:
- $\eta$: learning rate


```python
x = tf.constant(x_vals.reshape(-1, 1))  # shape (100, 1)
y = tf.constant(y_vals.reshape(-1, 1))  # shape (100, 1)

# Initialize weights
W = tf.Variable(tf.random.normal([1, 1]))  # shape (1, 1)
b = tf.Variable(tf.zeros([1]))             # shape (1,)
learning_rate = 0.1
epochs = 200

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + tf.exp(-z))

# Training loop
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        z = tf.matmul(x, W) + b  # shape (100, 1)
        y_pred = sigmoid(z)

        # Binary cross-entropy loss
        loss = -tf.reduce_mean(y * tf.math.log(y_pred + 1e-7) +
                               (1 - y) * tf.math.log(1 - y_pred + 1e-7))

    dW, db = tape.gradient(loss, [W, b])
    W.assign_sub(learning_rate * dW)
    b.assign_sub(learning_rate * db)

    if epoch % 20 == 0:
        print(f"Epoch {epoch:03d}: Loss = {loss.numpy():.4f}, W = {W.numpy()[0][0]:.4f}, b = {b.numpy()[0]:.4f}")

print(f"\nFinal Model: W = {W.numpy()[0][0]:.4f}, b = {b.numpy()[0]:.4f}")
```

```python
# Plot decision boundary
plt.scatter(x_vals, y_vals, label="Data")
x_test = np.linspace(0, 1, 100).reshape(-1, 1)
y_test = sigmoid(x_test @ W.numpy() + b.numpy()).numpy().flatten()
plt.plot(x_test, y_test, color='red', label="Logistic Regression Curve")
plt.title("Logistic Regression Fit")
plt.xlabel("x")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```
